```python
import pickle
from scipy.interpolate import interp1d
```

```python
a_file = open("BER_data_64QAM.pkl", "rb")
b_file = open('BER_KK_17_CSPR_64QAM.pkl','rb')
every_model_64QAM = pickle.load(a_file)
KK_17_cspr_64QAM = pickle.load(b_file)
a_file.close()
b_file.close()
```

```python
a_file = open("BER_model_2_sps_256_amostras_64QAM.pkl", "rb")
b_file = open("BER_KK_different_CSPRs_2_SpS_64QAM.pkl",'rb')
model2sps = pickle.load(a_file)
kk2sps = pickle.load(b_file)
```

```python
a_file = open("BER_model_4_sps_256_amostras_64QAM.pkl", "rb")
b_file = open("BER_KK_different_CSPRs_4_SpS_64QAM.pkl",'rb')
model4sps = pickle.load(a_file)
kk4sps = pickle.load(b_file)
```

```python
a_file = open("BER_model_8_sps_256_amostras_64QAM.pkl", "rb")
b_file = open("BER_KK_different_CSPRs_8_SpS_64QAM.pkl",'rb')
model8sps = pickle.load(a_file)
kk8sps = pickle.load(b_file)
```

```python
snr_ber_theo = np.arange(18,49,0.1)
SNRs = np.arange(18,49,2)
SpS = 8
M = 64
amostras = 512
#BERsAWGN = every_model_16QAM['BER_AWGN']
#f = interp1d(every_model_64QAM[f'BER_{SpS}_sps_{amostras}_amostras'],SNRs)
#penalty = np.round(f(1e-2) - snr_ber_theo[ber_theory(helpers.dB2lin(snr_ber_theo), M)<1e-2][0],2)

plt.plot(SNRs,every_model_64QAM[f'BER_{SpS}_sps_{amostras}_amostras'],'--P',c='r')
plt.plot(SNRs,KK_17_cspr_64QAM[f'BER_KK_{SpS}_sps'],'--p',c='b')
#plt.plot(SNRs,BERsAWGN,'-*',c='g')
plt.plot(snr_ber_theo,ber_theory(helpers.dB2lin(snr_ber_theo), M),'-g')
plt.axhline(y=1e-2,c='k',ls='--')
#plt.title(f'Model with {SpS} SpS - 16 QAM - {amostras} features')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.yscale('log')
plt.ylim([1.5e-3,1e-1])
plt.xlim([19,36])
plt.tight_layout()
plt.yticks([1e-3,1e-2,1e-1])
plt.legend()
plt.legend(['ANN','KK', 'BER teórica', 'limite FEC'])
plt.grid(True)
plt.savefig(f'BER_x_SNR_model_{SpS}_SpS_{amostras}_amostras_{M}QAM.png')
```

```python
SpS = 2
M = 64
if SpS == 2:
  model = model2sps
  kk = kk2sps
if SpS == 4:
  model = model4sps
  kk = kk4sps
if SpS == 8:
  model = model8sps
  kk = kk8sps
SNRs = np.arange(18,49,2)
snr_ber_theo = np.arange(18,49,0.1)
CSPR = 10
plt.plot(SNRs,model2sps[f'model_{SpS}_sps_256_amostras_{CSPR}_CSPR_64QAM'],'--^',label=f'ANN - {SpS} SpS')
plt.plot(SNRs,kk2sps[f'KK_{SpS}_sps_{CSPR}_CSPR'],'P',linestyle = (0, (1, 1)),label=f'KK - {SpS} SpS')
SpS = 4
plt.plot(SNRs,model4sps[f'model_{SpS}_sps_256_amostras_{CSPR}_CSPR_64QAM'],'--v',label=f'ANN - {SpS} SpS')
plt.plot(SNRs,kk4sps[f'KK_{SpS}_sps_{CSPR}_CSPR'],'X',linestyle = (0, (1, 1)),label=f'KK - {SpS} SpS')
SpS = 8
plt.plot(SNRs,model8sps[f'model_{SpS}_sps_256_amostras_{CSPR}_CSPR_64QAM'],'-->',label=f'ANN - {SpS} SpS')
plt.plot(SNRs,kk8sps[f'KK_{SpS}_sps_{CSPR}_CSPR'],'s',linestyle = (0, (1, 1)),label=f'KK - {SpS} SpS')

plt.plot(snr_ber_theo,ber_theory(helpers.dB2lin(snr_ber_theo), M),'-b',label ='BER teórica')
plt.axhline(y=1e-2,c='k',ls='--', label='limite FEC')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.yscale('log')
plt.ylim([1.5e-3,1e-1])
plt.xlim([19,36])
plt.tight_layout()
plt.yticks([1e-3,1e-2,1e-1])
plt.legend()
plt.grid(True)
plt.savefig(f'BER_x_SNR_model_256_amostras_{M}QAM_{CSPR}dB_CSPR_SpS_comparison.png')
```
